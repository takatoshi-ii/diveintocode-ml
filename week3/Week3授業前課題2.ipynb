## 【問題5】用語の調査

### ＜尖度＞

尖度(せんど):確率変数の確率密度関数や頻度分布の鋭さを表す数字である。  
             正規分布と比べて、尖度が大きければ鋭いピークと長く太い裾を持った分布を持ち、  
             尖度が小さければより丸みがかったピークと短く細い尾を持った分布であるという事が判断できる。

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,12))

img_sendo = mpimg.imread("./尖度.png")
imgplot = plt.imshow(img_sendo)
```

### ＜歪度＞

歪度(わいど):平均まわりの三次モーメントを標準偏差で正規化したもので，分布の歪み具合（どれくらい非対称なのか）を表す。

```python
plt.figure(figsize=(8,8))

img_waido = mpimg.imread("./歪度.png")
imgplot = plt.imshow(img_waido)
```

## 【問題6】分布の確認

```python
import seaborn as sns
import scipy
import numpy as np
sns.distplot(df_House.iloc[:,-1:], bins=50)


print("尖度：", scipy.stats.kurtosis(df_House.iloc[:,-1:]))
print("歪度：", scipy.stats.skew(df_House.iloc[:,-1:]))
print("平均：", np.mean((df_House.iloc[:,-1:])))
```

```python

df_log = df_House.iloc[:,-1:].apply(np.log)
df_log
sns.distplot(df_log.iloc[:,-1:], bins=50)
print("歪度", scipy.stats.skew(df_log))
print("尖度", scipy.stats.kurtosis(df_log))
```

```python
# 平均値
np.mean(df_log)
```

正規分布では、尖度・歪度 共に0となる。  
対数変換前の目的変数は尖度・歪度 共に高い値をしめしているが、  
対数変換を行う事によって尖度・歪度 共に0に近づき、正規分布に近い値となる。
