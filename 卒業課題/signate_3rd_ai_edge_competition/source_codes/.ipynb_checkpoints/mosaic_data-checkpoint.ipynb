```python
import os
import json
import numpy as np
import shutil
import random
import cv2 as cv

labels_in       = 'data_for_yolo_training.txt'
labels_out      = 'data_for_yolo_training_mosaic.txt'
out_dir         = 'imgs/mosaic/'
generate_images = 10000
out_file        = open(labels_out, "w")

os.makedirs(out_dir, exist_ok=True)


def is_intersect(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    if interArea > 0: return True
    return False


def iou_box(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou
    



with open(labels_in) as f:  lines    = f.readlines()
for i in range (0, len(lines)): lines[i] = lines[i].split()
f.close()

tot_imgs1 = len(lines)-1

gen = 0
while gen<generate_images:
    if gen % 200 ==1: print(gen)
    index1 = random.randint(0,tot_imgs1)
    index2 = random.randint(0,tot_imgs1)
        
        
    im1 = cv.imread(lines[index1][0])
    im2 = cv.imread(lines[index2][0])
    
    im = np.zeros(im1.shape).astype('uint8')
    max_x = im1.shape[1]
    max_y = im1.shape[0]
    half_x = random.randint(500,max_x-500)
    
    
    im[:, 0:half_x,:]         = im1[:, 0:half_x,:] 
    im[:, half_x:max_x,:]     = im2[:, half_x:max_x,:]
    
    areas = [np.asarray((0, 0, half_x, max_y)),
             np.asarray((half_x, 0, max_x, max_y))]
    indexes = [index1, index2]
    
    str_to_write = out_dir+str(gen)+'.jpg'
    for k in range (0, 2):
        boxes = np.array([np.array(list(map(int, box.split(',')))) for box in lines[indexes[k]][1:]])
        for b in range (0, len(boxes)):
            arr = boxes[b]
            if is_intersect(areas[k], np.asarray((arr[0],arr[1],arr[2],arr[3]))) == True:
                box = np.asarray((max(areas[k][0],arr[0]), max(areas[k][1],arr[1]), min(areas[k][2],arr[2]), min(areas[k][3],arr[3])))
                if (box[2]-box[0]) * (box[3]-box[1]) < 800:#we do not want to have small boxes
                    continue
                if arr[4] == 1: #pedestrian
                    if iou_box(box, np.asarray((arr[0],arr[1],arr[2],arr[3]))) < 0.3: #too small part, delete
                        continue
                else:
                    if iou_box(box, np.asarray((arr[0],arr[1],arr[2],arr[3]))) < 0.3: #too small part, delete
                        continue
                str_to_write += ' '+str(box[0])+','+str(box[1])+','+str(box[2])+','+str(box[3])+','+str(arr[4])
        
    if str_to_write != out_dir+str(gen)+'.jpg':
        out_file.write(str_to_write+'\n')
        
    cv.imwrite(out_dir+str(gen)+'.jpg', im)
    gen+=1
out_file.close()   
```
