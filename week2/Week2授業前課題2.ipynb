## 【問題1】乱数の作成

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]

# 2次元正規乱数を500個生成
values1 = np.random.multivariate_normal(mu, sigma, 500)
print("values1.shape:{}".format(values1.shape))
print("values1:\n{}".format(values1))
```

## 【問題2】散布図による可視化

```python
plt.scatter(values1[:,0], values1[:,1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.show()
```

```python
plt.xlim(-5, 3)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histgram of x1")
plt.hist(values1[:,0], bins = 50)
```

## 【問題4】データの追加

```python
mu = [0, -3]
sigma = [[1.0, 0.8], [0.8, 1.0]]

# 2次元正規乱数を500個生成
values2 = np.random.multivariate_normal(mu, sigma, 500)
print("values2.shape:{}".format(values2.shape))
print("values2[1:,0]={}".format(values2[1:,0]))
print("values2:\n{}".format(values2))
```

```python
plt.scatter(values1[:,0], values1[:,1],color= "blue", label ="0")
plt.scatter(values2[:,0], values2[:,1],color= "red", label ="1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

# 凡例の作成
plt.legend()
plt.show()
```

## 【問題5】データの結合

```python
values3 = np.concatenate([values1, values2], axis = 0)

print("values3.shape:{}".format(values3.shape))
print("values3:\n{}".format(values3))
```

## 【問題6】ラベル付け

```python
# 全要素が“0”の1次元配列を作成し、行列入れ替え
l1 = np.zeros(500).reshape(500,-1)

# 全要素が“1”の1次元配列を作成し、行列入れ替え
l2 = np.ones(500).reshape(500,-1)

# 行列の結合⇒1000x1へ
l3 = np.concatenate([l1, l2], axis = 0)

print("values3.shape:{}".format(l3.shape))
print("values3:\n{}".format(l3))

# 1000x2の行列へ列を結合⇒1000x3へ
values4 = np.concatenate([values3, l3], axis = 1)

print("values4.shape:{}".format(values4.shape))
print("values4:\n{}".format(values4))
```
