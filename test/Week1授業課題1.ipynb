# Week1授業課題1 曽呂利新左衛門問題

## 【問題1】100日目の米粒の数

```python
import matplotlib.pyplot as plt
%matplotlib inline

RICE= 1
DAYS = 100
rice_count = RICE
rice_ruikei = 0

rice_cnt_list = []     # 当日に貰える米の数
rice_ruikei_list = []  # 類型でもらえる米の数

for i in range(DAYS):
    rice_count *= 2
    rice_cnt_list.append(rice_count)
    rice_ruikei += rice_count
    rice_ruikei_list.append(rice_ruikei)
    

fig, ax1 = plt.subplots(facecolor="w")
ax1.set_title("sorori sinnzaemon")
ax1.set_xlabel("day")
ax1.set_ylabel("rice count", color ='blue')
ax1.grid()
ax1.plot(rice_cnt_list, color ='blue', linestyle='dashed')
ax1.tick_params(axis = 'y', colors ='blue')

ax2 = ax1.twinx()
ax2.set_yscale("log")
ax2.set_ylabel("rice count(log)", color ='red')
ax2.plot(rice_cnt_list,color='red')
ax2.tick_params(axis = 'y', colors ='red')

plt.show()
```

```python
# 貰った米の累計グラフを表示
fig, ax1 = plt.subplots(facecolor="w")
ax1.set_title("sorori sinnzaemon")
ax1.set_xlabel("day")
ax1.set_ylabel("rice count total", color ='blue')
ax1.grid()
ax1.plot(rice_ruikei_list, color ='blue', linestyle='dashed')
ax1.tick_params(axis = 'y', colors ='blue')

ax2 = ax1.twinx()
ax2.set_yscale("log")
ax2.set_ylabel("rice count(log)", color ='red')
ax2.plot(rice_ruikei_list,color='red')
ax2.tick_params(axis = 'y', colors ='red')

plt.show()
```

# 【問題2】100日目以外の米粒の数

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """

    RICE= 1    # 初期値は1粒
    rice_count = RICE
    rice_ruikei = RICE
    rice_cnt_list = [RICE]
    rice_ruikei_list = [RICE]

    for i in range(day):
        rice_count *= 2
        list_n_grains.append(rice_count)
        rice_ruikei += rice_count
        list_total_grains.append(rice_ruikei)
    
    return list_n_grains, list_total_grains


list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=120)
plt.title("sorori sinnzaemon")
plt.xlabel("day")
plt.ylabel("rice count")
plt.plot(rice_cnt_list, color="red") 
plt.plot(rice_ruikei_list, color = "blue") 
plt.show()
```

```python
print("list_n_grains =     " + str(list_n_grains))
print("list_total_grains = " + str(list_total_grains))
```

# 【問題3】何人が何日生活できる米か


```python
def life_sorori_shinzaemon(day, ninzu = 1, need_rice = 0):
    # 指定した日数で貰える米の数を算出
    list_n_grains, list_total_grains = compute_sorori_shinzaemon(day)
    
    # 貰えた米の累計数を人数と1日当たりの必要米粒で割る
    day = list_total_grains[-1] / (ninzu * need_rice)
    
    return day 
```

```python
# 茶碗一杯の米粒を3250粒とする
# 全ての人間が１食１杯だけ食べるとする。(ギャル曽根は対象外！)
# おかわり等は考慮しない
PRIOD = 120    # 4ヶ月(１２０日)
NINZU = 12     # １２人
NEED_RICE = 3250   # 3250粒/１杯
day = life_sorori_shinzaemon(PRIOD, NINZU, NEED_RICE*3)
print("{:.1f}日 生活が可能".format(day))
print("{:.1f}月 生活が可能".format(day/30))
print("{:.1f}年 生活が可能".format(day/30/12))
print("{:.1f}万年 生活が可能".format(day/30/12/10000))
```
